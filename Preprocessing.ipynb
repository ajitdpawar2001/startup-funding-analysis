#Preprocessing:
- Drop remarks col
- Set index sno
- Rename col
- Convert amount to cr RS
- Date col
- Dropna
- Separate all investors


```python
df.head(2)
```

```python
df.drop(columns=['Remarks'],inplace=True)
```

```python
df.head(2)
```

```python
df.set_index('Sr No',inplace=True)
```

```python
df.head(1)
```

```python
df.rename(columns={
  'Date dd/mm/yyyy':'date',
   'Startup Name':'startup',
   'Industry Vertical':'vertical',
   'SubVertical':'subvertical',
   'City  Location':'city',
   'Investors Name':'investors',
   'InvestmentnType':'round',
   'Amount in USD':'amount'
},inplace=True)
```

```python
df.head(1)
```

```python
df['amount']=df['amount'].fillna('0')
```

```python
df['amount'].info()
```

```python
#df['amount'].astype('int')   #it gives error
df['amount']=df['amount'].str.replace(',','')
```

```python
# df['amount'].astype('int')
df['amount']=df['amount'].str.replace('undisclosed','0')
```

```python
#df['amount'].astype('int')
df['amount']=df['amount'].str.replace('unknown','0')
```

```python
#df['amount'].astype('int')
df['amount']=df['amount'].str.replace('Undisclosed','0')
```

```python
#df['amount'].astype('int')
df=df[df['amount'].str.isdigit()]
```

```python
df['amount']=df['amount'].astype('float')
```

```python
def to_inr(dollar):
  inr = dollar * 88.25
  return inr

```

```python
df['amount'] = df['amount'].apply(to_inr)
```

```python
df['amount']
```

```python
#CONVERT DATE to DATE proper
df['date']= pd.to_datetime(df['date'] ,  errors = 'coerce')
```

```python
df['date']
```

```python
df.columns
```

```python
df.dropna(subset = [ 'date' , 'startup' , 'vertical' , 'city' , 'investors' , 'round' , 'amount'],inplace =True)
```

```python
df.to_csv('cleaned_data.csv' , index = False)
```
